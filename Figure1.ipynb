```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'Arial'
```

# Figure 1d

```python
colors = ['#e73847', '#f0faef', '#a8dadb', '#457b9d', '#1d3557']
e1s = [4.758, 6.443, 4.364, 3.960]
e2s = [3.665, 6.775, 4.292, 3.922]
r1s = [0.401, 0.666, 0.313, 0.283]
r2s = [0.488, 1.148, 0.649, 0.569]
refs = [5.189, 5.373, 3.761, np.nan]
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
lag = 3.5

plt.barh([0, lag, lag*2, lag*3], e1s, xerr = r1s, capsize = 5, label = 'Average Speed Model', color = colors[3], alpha = 0.6)
plt.barh([1, lag+1, lag*2+1, lag*3+1], e2s, xerr = r2s, capsize = 5, label = 'Instantaneous Model', color = colors[2], alpha = 0.5)
plt.barh([2, lag+2, lag*2+2, lag*3+2], refs, label = 'Carbon Monitor', color = colors[0], alpha = 0.5)
plt.legend(frameon = False, fontsize = 18, loc = 'upper right',)
plt.xlabel('Total Emisison (kt)', size = 18)
plt.xticks(size = 16)
plt.yticks([1, lag+1, lag*2+1, lag*3+1], ['S-1', 'S-2', 'J-1', 'J-2'], size = 16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.invert_yaxis()
plt.xlim(0, 8)
plt.tight_layout()
plt.show()
```

# Figure 1e

```python
Urs = np.array([6.9, 6.9, 3.51, 3.51])
Uts = np.array([0.2, 0.18, 0.08, 0.05])
Uss = np.array([5.84, 9.12, 5.26, 5.09])
Um1s = np.array([4.84, 7.68, 6.27, 6.22])
Um2s = np.array([9.78, 12.50, 13.75, 13.13])
U1s = np.array([8.43, 10.33, 7.18, 7.14])
U2s = np.array([13.32, 16.94, 15.13, 14.51])

mldcolors = ['#B4C7E7', '#C4B7D8', '#B7D9A5', '#D9AFCD', '#D9D9D9']
```

```python
fig, ax = plt.subplots(figsize = (9, 6))
lag = 3.5

plt.barh([0, lag, lag*2, lag*3, 1, lag+1, lag*2+1, lag*3+1], np.array([Urs, Urs]).reshape(-1), color = mldcolors[0], label = '$U_r$')
plt.barh([0, lag, lag*2, lag*3, 1, lag+1, lag*2+1, lag*3+1], np.array([Uts, Uts]).reshape(-1), left = np.array([Urs, Urs]).reshape(-1), color = mldcolors[1], label = '$U_t$')
plt.barh([1, lag+1, lag*2+1, lag*3+1], Uss, left = Urs + Uts, color = mldcolors[4], label = '$U_s$')

plt.barh([0, lag, lag*2, lag*3], Um1s, left = Urs + Uts, color = mldcolors[2], label = '$U_{m,1}$')
plt.barh([1, lag+1, lag*2+1, lag*3+1], Um2s, left = Urs + Uts + Uss, color = mldcolors[3], label = '$U_{m,2}$')

for i in range(4):
    plt.text(x = Urs[i] + Uts[i] + Um1s[i] + 0.2, y = lag * i, s = "$U_1$=%.2f%%" % (U1s[i]), 
            horizontalalignment = 'left', verticalalignment = 'center', fontsize = 13)
    plt.text(x = Urs[i] + Uts[i] + Uss[i] + Um2s[i] + 0.2, y = lag * i + 1, s = "$U_2$=%.2f%%" % (U2s[i]), 
            horizontalalignment = 'left', verticalalignment = 'center', fontsize = 13)

plt.legend(frameon = False, fontsize = 15, loc = 'lower right',)
plt.xlabel('Uncertainty (%)', size = 18)
plt.xticks(size = 16)
plt.yticks([])
plt.xlim(0, 33)

plt.yticks([0.5, lag+0.5, lag*2+0.5, lag*3+0.5], ['S-1', 'S-2', 'J-1', 'J-2'], size = 16)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```
